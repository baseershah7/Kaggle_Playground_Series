```python
train_org = pd.read_csv('/kaggle/input/uhi-index-data/Training_data_uhi_index_UHI2025-v2.csv')
train = pd.read_csv('/kaggle/input/ey-peak-0-9791/ey_train_niche_ds_0.9791.csv')
test  = pd.read_csv('/kaggle/input/ey-peak-0-9791/ey_test_niche_ds_0.9791.csv')
train['UHI Index'] = train_org['UHI Index']
```

```python
import h2o
from h2o.automl import H2OAutoML

# 1. Restart H2O and train fresh
h2o.init()  # Start new session

# 2. Convert data AFTER restart
h2o_df = h2o.H2OFrame(train)  # Recreate H2OFrame in new session
y = 'UHI Index'
X = [col for col in train.columns if col != 'UHI Index']
# 3. Define AutoML and retrain
aml = H2OAutoML(
    max_models=50,
    max_runtime_secs=39600,
    seed=42,
    sort_metric='r2',
    nfolds=5,
    keep_cross_validation_predictions=True,
    include_algos=["XGBoost", "DeepLearning", "StackedEnsemble", "GBM", "DRF", "GLM"]
)

aml.train(x=X, y=y, training_frame=h2o_df)  # Retrain in new session
print(h2o.ls())
# 4. Now access leader model
lb = aml.leaderboard
print(lb)

# Verify leader exists
if aml.leader is not None:
    leader_model = aml.leader
    oof_predictions = leader_model.cross_validation_holdout_predictions()
    print(oof_predictions.head())
else:
    print("No leader model found - check training logs!")
```

```python
lb = lb.as_data_frame()  
def get_all_oof_preds(aml):

    lb = aml.leaderboard.as_data_frame()
    model_ids = lb['model_id'].tolist()
    
    oof_dfs = []
    for mid in model_ids:
        try:
            model = h2o.get_model(mid)
            oof_preds = model.cross_validation_holdout_predictions().as_data_frame(use_multi_thread=False)
            oof_preds.rename(columns={'predict': mid}, inplace=True)
            oof_dfs.append(oof_preds)
        except:
            print(f"Skipping {mid} - no OOF predictions available")
    
    return pd.concat(oof_dfs, axis=1)

all_oof_preds = get_all_oof_preds(aml)

print("\nPrediction DataFrame Info:")
print(f"Shape: {all_oof_preds.shape}")
print(f"Columns: {list(all_oof_preds.columns)}")
print("\nSample predictions:")
print(all_oof_preds.head())

print("\nLeaderboard:")
print(aml.leaderboard)
```

```python
all_oof_preds.to_csv('oof_preds_h2o.csv', index=False)
```

```python
leaderboard = aml.leaderboard
model_save_dir = '/kaggle/working/raw_data_models_final/'  

for model_id in leaderboard['model_id'].as_data_frame()['model_id']:
    model = h2o.get_model(model_id)
    model_path = h2o.save_model(model=model, path=model_save_dir, force=True)
    print(f"Model {model_id} saved to {model_path}")

leaderboard_df = leaderboard.as_data_frame()
leaderboard_csv_path = os.path.join(model_save_dir, 'leaderboard.csv')
leaderboard_df.to_csv(leaderboard_csv_path, index=False)
print(f"Leaderboard saved to {leaderboard_csv_path}")
h2o.shutdown()
```
